```python
import os 
import sys
import numpy as np
import matplotlib.pyplot as plt

current_dir = os.getcwd()
parent_dir = os.path.join(current_dir, "..")
sys.path.append(parent_dir)

from utils import tck_to_tdi
```

```python
TRACTS_10_DIR = '../HCP10_Tracts'
MRI_10_DIR = '../HCP10_MRI'
TEMPLATE_SUBJECT = '959574'
SUBJECT_IDS = list(filter(lambda x: not x.endswith('.txt') and x != TEMPLATE_SUBJECT, os.listdir(TRACTS_10_DIR)))
    
subject_id = SUBJECT_IDS[0]
warped_tck_path = os.path.join("../output", "mrregister1", subject_id, f"{subject_id}_CC_warped.tck")
groundtruth_tck_path = os.path.join(TRACTS_10_DIR, subject_id, "tracts", "CC.tck")
#reference_mri_path = os.path.join(MRI_10_DIR, "data", f"{TEMPLATE_SUBJECT}_Diffusion3TRecommended", f"{TEMPLATE_SUBJECT}", "T1w", "Diffusion", "data.nii.gz")
reference_mri_path = os.path.join(MRI_10_DIR, 'data', f"{TEMPLATE_SUBJECT}_StructuralRecommended", f"{TEMPLATE_SUBJECT}", 'T1w', 'T1w_acpc_dc_restore_brain.nii.gz')
```

```python
warped_tdi = tck_to_tdi(warped_tck_path, reference_mri_path)
groundtruth_tdi = tck_to_tdi(groundtruth_tck_path, reference_mri_path)
```

```python
axial_slice_idx = warped_tdi.shape[2] // 2
#non_zero_tdi = warped_tdi[warped_tdi > 0]
#robust_vmax = np.percentile(non_zero_tdi, 99)

plt.figure(figsize=(8,8))

plt.imshow(warped_tdi[:, :, axial_slice_idx].T, cmap='hot', origin='lower')

plt.colorbar(label='Streamline Density (Count)')

plt.title(f'TDI Axial Slice (Z = {axial_slice_idx})')
plt.axis('off')
plt.show()
```

```python
axial_slice_idx = warped_tdi.shape[2] // 2
combined_min = min(warped_tdi.min(), groundtruth_tdi.min())
combined_max = max(warped_tdi.max(), groundtruth_tdi.max())
figure = plt.figure(figsize=(8,8))
cols, rows = 3, 1
for i in range(1, cols * rows + 1):
    if i == 1:
        img = warped_tdi[:, :, axial_slice_idx]
    if i == 2:
        img = groundtruth_tdi[:, :, axial_slice_idx]
    if i == 3:
        img = groundtruth_tdi[:, :, axial_slice_idx] - warped_tdi[:, :, axial_slice_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(f'TDI Axial Slice (Z = {axial_slice_idx})')
    plt.axis("off")
    plt.imshow(img.T, cmap='hot', origin='lower')
    plt.colorbar(label='Streamline Density (Count)')

plt.axis('off')
```

```python
axial_slice_idx = warped_tdi.shape[2] // 2

# 1. Determine Shared Range for the first two images
combined_min = min(warped_tdi.min(), groundtruth_tdi.min())
combined_max = max(warped_tdi.max(), groundtruth_tdi.max())

# 2. Determine Symmetric Range for the difference map
difference_data = groundtruth_tdi[:, :, axial_slice_idx] - warped_tdi[:, :, axial_slice_idx]
max_abs_diff = np.abs(difference_data).max()

# Create the figure and axes objects
fig, axes = plt.subplots(1, 3, figsize=(15, 6)) # Adjusted figsize for better viewing

# --- Subplot 1: Warped TDI ---
img1 = warped_tdi[:, :, axial_slice_idx]
ax1 = axes[0]
im1 = ax1.imshow(img1.T, cmap='hot', origin='lower',
                 vmin=combined_min, vmax=combined_max) # <--- Apply Shared Range
ax1.set_title(f'Warped TDI (Z = {axial_slice_idx})')
ax1.axis("off")
fig.colorbar(im1, ax=ax1, label='Streamline Density (Count)')

# --- Subplot 2: Groundtruth TDI ---
img2 = groundtruth_tdi[:, :, axial_slice_idx]
ax2 = axes[1]
im2 = ax2.imshow(img2.T, cmap='hot', origin='lower',
                 vmin=combined_min, vmax=combined_max) # <--- Apply Shared Range
ax2.set_title(f'Groundtruth TDI (Z = {axial_slice_idx})')
ax2.axis("off")
fig.colorbar(im2, ax=ax2, label='Streamline Density (Count)')

# --- Subplot 3: Difference Map ---
img3 = difference_data
ax3 = axes[2]
# Use a diverging colormap (like 'RdBu' or 'coolwarm') and center it at zero
im3 = ax3.imshow(img3.T, cmap='hot', origin='lower',
                 vmin=combined_min, vmax=combined_max) # <--- Apply Symmetric Range
ax3.set_title('Difference (Groundtruth - Warped)')
ax3.axis("off")
fig.colorbar(im3, ax=ax3, label='Density Difference (Count)')

plt.tight_layout()
plt.show()
```

```python
def plot_subject_comparison(subject_id, warped_tdi, groundtruth_tdi, save_path =None):
    """
    Plots the warped TDI, groundtruth TDI, and their difference for a single subject.
    """
    # Use the central axial slice
    axial_slice_idx = warped_tdi.shape[2] // 2

    warped_slice = warped_tdi[:, :, axial_slice_idx]
    groundtruth_slice = groundtruth_tdi[:, :, axial_slice_idx]
    difference_data = groundtruth_slice - warped_slice

    # --- Determine Ranges ---
    # 1. Shared Range for the two original images
    combined_min = min(warped_tdi.min(), groundtruth_tdi.min())
    combined_max = max(warped_tdi.max(), groundtruth_tdi.max())

    # 2. Symmetric Range for the difference map
    max_abs_diff = np.abs(difference_data).max()
    
    # --- Create Figure ---
    fig, axes = plt.subplots(1, 3, figsize=(15, 6))
    fig.suptitle(f'Subject {subject_id} TDI Comparison (Z = {axial_slice_idx})', 
                 fontsize=16, y=1.02)

    # --- Subplot 1: Warped TDI ---
    ax1 = axes[0]
    im1 = ax1.imshow(warped_slice.T, cmap='hot', origin='lower',
                     vmin=combined_min, vmax=combined_max)
    ax1.set_title('Warped TDI')
    ax1.axis("off")
    fig.colorbar(im1, ax=ax1, label='Streamline Density (Count)')

    # --- Subplot 2: Groundtruth TDI ---
    ax2 = axes[1]
    im2 = ax2.imshow(groundtruth_slice.T, cmap='hot', origin='lower',
                     vmin=combined_min, vmax=combined_max)
    ax2.set_title('Groundtruth TDI')
    ax2.axis("off")
    fig.colorbar(im2, ax=ax2, label='Streamline Density (Count)')

    # --- Subplot 3: Difference Map ---
    ax3 = axes[2]
    # Use 'coolwarm' for diverging map, centered at zero
    im3 = ax3.imshow(difference_data.T, cmap='hot', origin='lower',
                     vmin=combined_min, vmax=combined_max)
    ax3.set_title('Difference (Groundtruth - Warped)')
    ax3.axis("off")
    fig.colorbar(im3, ax=ax3, label='Density Difference (Count)')

    plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout for suptitle
    #plt.show() # Use this if you want figures to appear sequentially
    # Or, fig.savefig(f'{subject_id}_comparison.png') to save them

    if save_path:        
        # Save the figure
        fig.savefig(save_path, dpi=300, bbox_inches='tight') # Saves with high resolution
        
        # Close the figure to free up memory (important in a large loop)
        plt.close(fig)

        print(f"Saved plot for {subject_id} to {filename}")

    else:
        plt.show()
```

```python
for subject_id in SUBJECT_IDS:
    warped_tck_path = os.path.join("../output", "mrregister1", subject_id, f"{subject_id}_CC_warped.tck")
    groundtruth_tck_path = os.path.join(TRACTS_10_DIR, subject_id, "tracts_tck", "CC.tck")
    save_path = os.path.join("../output", "mrregister1", subject_id, f'{subject_id}_TDI_Comparison.png')
    warped_tdi = tck_to_tdi(warped_tck_path, reference_mri_path)
    groundtruth_tdi = tck_to_tdi(groundtruth_tck_path, reference_mri_path)
    
    plot_subject_comparison(
        subject_id, 
        warped_tdi, 
        groundtruth_tdi,
        save_path
    )
```
